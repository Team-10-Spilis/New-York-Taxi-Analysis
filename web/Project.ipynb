```python
data = pd.read_parquet("final_sample.parquet")
```

```python
data = data.drop(labels = [
    "source_file",
    "sample_year",
    "sample_month",
    "month_total_rows",
    "month_sample_rows",
    "target_sample_frac",
    "actual_sample_frac",
    "cap_was_applied",
], axis = 1)
```

```python
iterator = data.iterrows()
```

```python
for index, row in iterator:
  if (str(row['airport_fee']) == "nan"):
    if (str(row['Airport_fee']) != "nan"):
      row['airport_fee'] = row['Airport_fee']
    else:
      row['airport_fee'] = 0
```
